```python
# python 多线程卖票 threading类实例化
import time, random, threading

lock = threading.Lock()
tickets = 100

def run(name):
    global lock
    global tickets

    while tickets:
        lock.acquire()
        if tickets > 0:
            tickets -= 1
            print(f'{name}卖出一张,剩余{tickets}张票')
        lock.release()
        time.sleep(random.randint(0, 2))

def main():
    threads = {}
    for i in ('A', 'B'):
        j = threading.Thread(target=run, args=(i, ))
        threads[i] = j

    # for i in ('A', 'B'):
    #     # 开始执行多线程
    #     threads[i].start()

    threads['A'].start()
    threads['B'].start()

    # for i in ('A', 'B'):
    #     # 等待多线程结束往下执行
    #     threads[i].join()

    threads['A'].join()
    threads['B'].join()

    print('exiting main thread')

if __name__ == '__main__':
    main()
```

```python
# python 多线程卖票 thread集成
import time, random, threading

tickets = 100

class myThread(threading.Thread):
    
    def __init__(self, threadid, name):
        threading.Thread.__init__(self)
        self.threadid = threadid
        self.name = name

    def run(self):
        global tickets
        while tickets:
            threadLock = threading.Lock()
            threadLock.acquire()
            if tickets > 0:
                tickets -= 1
                print(f'{self.name}卖出一张,剩余{tickets}张票')
            threadLock.release()
            time.sleep(random.randint(0, 2))



thread1 = myThread(1, 'A')
thread2 = myThread(2, 'B')

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print('exiting main thread')
```
